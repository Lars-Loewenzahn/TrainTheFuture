### Höhe eine Baumes

$\alpha$ = 19

```python
import matplotlib.pyplot as plt

jahre =  list(range(11))
hoehen = []
for zeit in jahre:
    hoehe = 80 + 25*zeit
    hoehen.append(hoehe)

plt.scatter(jahre, hoehen, marker='x')
plt.title("Die Höhe eines Baumes in Jahren")
plt.xlabel("Zeit in Jahren")
plt.ylabel("Höhe in cm")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

theta = np.linspace(0, np.pi/2, 100)
r = 1
x = r * np.cos(theta)
y = r * np.sin(theta)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.fill_between(x, y, color='lightblue')
ax.set_aspect('equal')
plt.title("1/4 Kreis (Viertelkreis)")
plt.show()
```
