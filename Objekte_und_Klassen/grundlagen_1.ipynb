```python
class Schueler:
    def __init__(self, n, a):
        self.name = n
        self.alter = a
    
    def vorstellen(self):  # Methoden sind Funktionen einer Klasse
        return "Hallo, ich bin " + self.name + " und ich bin " + str(self.alter) + " Jahre alt"

    def meinung(self, fach):
        return "Ich finde " + fach + " doof!"

    def dialog(self, other):
        ans = "Hallo ich bin " + self.name + ", wer bist du? \n"
        ans += "Hallo, mein Name ist " + other.name + "!"
        return ans

s1 = Schueler("Lars", 32)
s2 = Schueler("Lina", 19)
```

```python
print(s1.dialog(s2))
```

```python
class Cat: 
    def __init__(self , name):
        self.name = name
        self.happiness = 0

    def printCat(self):
        print(f"Meow, ich bin {self.name}, und ich {self.happiness}/10 Gl端cklich!")
    
    def meow(self):
        print("Meow!")

    def pet(self):
        self.happiness += 1
        self.meow()
        self.check_happiness()
    
    def annoy(self):
        self.happiness -= 1
        self.check_happiness()
    
    def check_happiness(self):
        if self.happiness > 10:
            self.happiness = 10
            print(" **schnurr** ")
        if self.happiness < 0:
            self.happiness = 0
            print(" **Fauch** ")
```

```python
c1 = Cat("Jerry")
print(c1.name, c1.happiness)
c1.printCat()
```
